```python
df2['num_accounts_120d_past_due'].value_counts()
```

```python
columns_to_impute = ['emp_length', 'debt_to_income', 'months_since_last_delinq',
                     'months_since_90d_late', 'months_since_last_credit_inquiry']

# Perform median imputation for each column
for column in columns_to_impute:
    median_value = df2[column].median()
    df2[column].fillna(median_value, inplace=True)

df2['emp_title'] = df2['emp_title'].fillna('unemployed')
df2['num_accounts_120d_past_due'] = df2['num_accounts_120d_past_due'].fillna(0)
```

```python
df2['loan_status'].value_counts()
```

```python
df2['loan_status'].replace({'In Grace Period': 'Late', 'Late (31-120 days)': 'Late', 'Late (16-30 days)': 'Late'}, inplace=True)
df2['loan_status'] = (df2['loan_status'] == 'Charged Off').astype(int)
```

```python
df2.to_parquet('../data/loans_full_schema_clean.parquet', index=None)
```
